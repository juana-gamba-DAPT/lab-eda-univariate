```python
#tabla de frecuencias

import pandas as pd

df = pd.read_csv('C:/Users/juana/OneDrive/Escritorio/ironhack/lab-eda-univariate/archive (3)/amz_uk_price_prediction_dataset.csv') 

# Generar tabla de frecuencia
category_counts = df['category'].value_counts()

# Mostrar las 5 más comunes
top_5_categories = category_counts.head(5)
print(top_5_categories)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Barplot para las 10 categorías más frecuentes
plt.figure(figsize=(12,6))
sns.barplot(x=top_5_categories.values, y=top_5_categories.index, palette='viridis')
plt.title('Top 5 Product Categories by Frequency')
plt.xlabel('Number of Listings')
plt.ylabel('Product Category')
plt.show()
```

```python
# Pie chart solo para las top 5
plt.figure(figsize=(8,8))
plt.pie(top_5_categories, labels=top_5_categories.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Proportion of Top 5 Product Categories')
plt.axis('equal')
plt.show()
```
